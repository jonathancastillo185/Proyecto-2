```python
df = pd.read_csv(r'..\Data\Internet_Accesos-por-velocidad.csv')
```

```python
df.drop(columns=['OTROS'],inplace=True)
```

```python
df
```

```python

df = df[df['Provincia'] == 'Buenos Aires']

plt.boxplot(df['Total'])

plt.title('Distribución del acceso a Internet en Buenos Aires')
plt.ylabel('Acceso a Internet')

plt.show()

```

```python
BAF = pd.read_csv(r'..\Data\Internet_BAF.csv')
```

```python
BAF
```

```python

df['diferencia'] = df.groupby('Provincia')['Total'].diff()
```

```python

#df['diferencia'] = round(df['diferencia'],2)

df = df.sort_values(by='Año', ascending=True)
```

```python
df
```

```python
# Crear el gráfico de barras

plt.figure(figsize=(12, 8))

sns.barplot(x=df['Provincia'], y=df['diferencia'])

plt.xticks(rotation=45)

plt.show()
```

```python


plt.figure(figsize=(8, 6))  # Ancho x Alto

# Crear el gráfico de líneas
sns.lineplot(x=df.loc[df['Provincia'] == 'Capital Federal'][['Total']])

# Títulos y etiquetas
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
